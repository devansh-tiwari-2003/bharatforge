```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json("/content/DataEngineeringQ2.json")
df
```

```python
# Explode the patientDetails column
df_exploded = df.explode('patientDetails')

# Normalize the patientDetails column
df_normalized = pd.json_normalize(df_exploded['patientDetails'])

# Check if df_normalized is empty and print a warning if so
if df_normalized.empty:
    print("'patientDetails' column is empty, resulting in an empty normalized DataFrame.")

# Concatenate the original DataFrame with the normalized patientDetails
df_final = pd.concat([df_exploded.reset_index(drop=True), df_normalized], axis=1)

# If df_normalized is not empty, proceed with calculating missing values
if not df_normalized.empty:
    # Calculate the percentage of missing values
    missing_percentage = df_final[['firstName', 'lastName', 'DOB']].isnull().sum() / len(df_final) * 100

    print(missing_percentage)
else:
    print("Cannot calculate missing values due to empty 'patientDetails' column.")
```

```python
# Calculate the mode of the gender column
mode_gender = df['patientDetails'].apply(lambda x: x.get('gender')).mode()[0]

# Fill missing gender values with the mode
df['patientDetails'] = df['patientDetails'].apply(lambda x: {**x, 'gender': x.get('gender', mode_gender)})

# Calculate the percentage of female gender
female_percentage = (df['patientDetails'].apply(lambda x: x['gender'] == 'female').sum() / len(df)) * 100

print(female_percentage)
```

```python
# Define a function to categorize age groups
def categorize_age(age):
    if age <= 12:
        return 'Child'
    elif 13 <= age <= 19:
        return 'Teen'
    elif 20 <= age <= 59:
        return 'Adult'
    else:
        return 'Senior'

# Extract age from patientDetails, handle missing ages, and apply the categorization function
df['ageGroup'] = df['patientDetails'].apply(lambda x: categorize_age(x.get('age', 0))) # Use get() with a default value to handle missing 'age' keys

# Count the number of adults, handling the potential absence of 'Adult'
adult_count = df['ageGroup'].value_counts().get('Adult', 0) # Use get() with a default value to handle missing 'Adult' key
print(adult_count)
```

```python
# Assuming 'consultationData' contains a list of prescribed medicines for each appointment
df['num_medicines'] = df['consultationData'].apply(lambda x: len(x.get('medicines', [])))
average_medicines = df['num_medicines'].mean()
print(average_medicines)
```

```python
# Calculate the count of active and inactive medicines
active_count = df['num_medicines'].sum()
total_count = len(df) * 4  # Assuming maximum 4 medicines per person
inactive_count = total_count - active_count

# Calculate percentages
active_percentage = (active_count / total_count) * 100
inactive_percentage = (inactive_count / total_count) * 100

# Print results
print("Percentage of active medicines: {:.2f}%".format(active_percentage))
print("Percentage of inactive medicines: {:.2f}%".format(inactive_percentage))
```

```python
def is_valid_indian_number(number):
    # Check if the number starts with 6, 7, 8, or 9 and has 10 digits.
    return number.startswith(('6', '7', '8', '9')) and len(number) == 10

# Apply the function to the 'phoneNumber' column and create a new column 'isValidMobile'
df['isValidMobile'] = df['phoneNumber'].apply(is_valid_indian_number)
```

```python
def is_valid_number(number):
    # Remove any non-digit characters
    number = ''.join(filter(str.isdigit, number))

    # Check prefix and range
    return (number.startswith('91') or number.startswith('+91')) and 6000000000 <= int(number) <= 9999999999

# Apply the function to the 'phoneNumber' column and count valid numbers
valid_count = df['phoneNumber'].apply(is_valid_number).sum()

print(valid_count)  # Output: Number of valid phone numbers
```
